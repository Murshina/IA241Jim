# The 10 Countries With Lowest Number of Cases

```python
sum_cases_by_country = df.groupby('countriesAndTerritories').sum()['cases']
sum_cases_by_country.nsmallest(10).plot.bar()
```

Less developed countries have the lowest number of cases due to limited testing 

# Average Number of Deaths Per Day

```python
mean_death_by_country = df.groupby('day').mean()['deaths']
df.plot.scatter(x='day',y='deaths')
```

Although hard to read, visualization shows majority of countries are experiencing a low number of deaths.

# Average Number of Cases Per Day

```python
mean_cases_by_country = df.groupby('dateRep').mean()['cases']
mean_cases_by_country.plot()
```

Spikes in the data does not necessarily suggest that cases are increasing, but that testing has increased.
